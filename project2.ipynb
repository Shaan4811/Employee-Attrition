```python
data['YearsAtCompany'].unique()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='OverTime',hue='Attrition',data=data)
plt.title('Does Overtime Effect Attrition');
```

it clearly means that people doing overtime also want to leave the comapany,so overtime effect attrition

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='JobRole', hue='Attrition',data=data,palette='Set1')
plt.title('Attrition by JobRole')
plt.xticks(rotation=45);
```

Attrition on the basis of Jobrole

```python
plt.figure(figsize=(8, 6))
attritionbydept= data.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()* 100
attritionbydept.plot(kind='bar', stacked=True, title='Attrition by Department');
```

Attrition on the basis of departments.

```python
plt.figure(figsize=(10, 6))
sns.countplot(hue='Attrition',x='Age',data=data,palette='pastel')
plt.title('Age Distribution by Attrition')
plt.grid()
plt.show()
```

Attrition on the basis of age

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='BusinessTravel', hue='Attrition',data=data,palette='Set1')
plt.title('Attrition By Businees Travel')
plt.grid();
```

here attrition rate depends on the travellers,some travel has the highest chance for attrition

```python
plt.figure(figsize=(8, 6))
sns.countplot(hue='Attrition', x='YearsSinceLastPromotion', data=data,palette='deep')
plt.title('Years Since Last Promotion by Attrition Status');
```

shoes that frequently promoted people are more willing to leave the company

```python
plt.figure(figsize=(8,6))
sns.countplot(hue='Attrition',x='DistanceFromHome (KM)',data=data)
plt.xticks(rotation=90)
plt.title('Attrition On The Basis Of DistanceFromHome');
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='YearsAtCompany',hue='Attrition',data=data,palette='muted')
plt.title('Attrition Rate On The Basis Of YearsAtCompany');
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='YearsWithCurrManager',hue='Attrition',data=data,palette='dark')
plt.title('Attrition On The Basis Of YearsWithCurrManager ')
```
